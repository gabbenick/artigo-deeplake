```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances

# --- 1. DEFINA O ARQUIVO A SER ANALISADO ---
# Altere para o arquivo .npz que deseja analisar.
embeddings_file_path = 'experiment_results/embeddings_from_deeplake_model.npz'

print(f"Analisando o arquivo: {embeddings_file_path}\n")

# --- 2. CARREGUE OS DADOS DO ARQUIVO .NPZ ---
try:
    data = np.load(embeddings_file_path)
    
    # --- CORREÇÃO AQUI ---
    # Usando os nomes que você descobriu.
    embedding_key = 'embeddings' 
    label_key = 'labels'
    
    test_embeddings = data[embedding_key]
    test_labels = data[label_key]
    print(f"Carregados {len(test_embeddings)} embeddings e rótulos.")

except FileNotFoundError:
    print(f"ERRO: O arquivo '{embeddings_file_path}' não foi encontrado.")
    exit()
except KeyError:
    print(f"ERRO: Chaves incorretas. Verifique os nomes em 'data.files' e ajuste no script.")
    exit()

# --- 3. CÁLCULO DAS ESTATÍSTICAS DE DISTÂNCIA ---
print("\nIniciando cálculo das distâncias...")
embeddings_class_0 = test_embeddings[test_labels == 0]
embeddings_class_1 = test_embeddings[test_labels == 1]

print("Calculando distâncias intra-classe...")
dist_intra_0 = euclidean_distances(embeddings_class_0) 
dist_intra_1 = euclidean_distances(embeddings_class_1)
intra_distances = np.concatenate([
    dist_intra_0[np.triu_indices(len(dist_intra_0), k=1)],
    dist_intra_1[np.triu_indices(len(dist_intra_1), k=1)]
])

print("Calculando distâncias inter-classe...")
dist_inter = euclidean_distances(embeddings_class_0, embeddings_class_1)
inter_distances = dist_inter.flatten()

stats_intra = {'Média': np.mean(intra_distances), 'Desvio Padrão': np.std(intra_distances)}
stats_inter = {'Média': np.mean(inter_distances), 'Desvio Padrão': np.std(inter_distances)}

# --- 4. EXIBIÇÃO DA TABELA DE RESULTADOS ---
print("\n" + "="*65)
print(f"  Análise de Distâncias para o Modelo: {embeddings_file_path.split('/')[-1]}")
print("="*65)
print(f"{'Tipo de Distância':<20} | {'Média':<20} | {'Desvio Padrão':<20}")
print("-" * 65)
print(f"{'Intra-classe':<20} | {stats_intra['Média']:<20.4f} | {stats_intra['Desvio Padrão']:<20.4f}")
print(f"{'Inter-classe':<20} | {stats_inter['Média']:<20.4f} | {stats_inter['Desvio Padrão']:<20.4f}")
print("="*65)
```

```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances

# --- 1. DEFINA O ARQUIVO A SER ANALISADO ---
# Altere para o arquivo .npz que deseja analisar.
embeddings_file_path = 'experiment_results/embeddings_from_png_model.npz'

print(f"Analisando o arquivo: {embeddings_file_path}\n")

# --- 2. CARREGUE OS DADOS DO ARQUIVO .NPZ ---
try:
    data = np.load(embeddings_file_path)
    
    # --- CORREÇÃO AQUI ---
    # Usando os nomes que você descobriu.
    embedding_key = 'embeddings' 
    label_key = 'labels'
    
    test_embeddings = data[embedding_key]
    test_labels = data[label_key]
    print(f"Carregados {len(test_embeddings)} embeddings e rótulos.")

except FileNotFoundError:
    print(f"ERRO: O arquivo '{embeddings_file_path}' não foi encontrado.")
    exit()
except KeyError:
    print(f"ERRO: Chaves incorretas. Verifique os nomes em 'data.files' e ajuste no script.")
    exit()

# --- 3. CÁLCULO DAS ESTATÍSTICAS DE DISTÂNCIA ---
print("\nIniciando cálculo das distâncias...")
embeddings_class_0 = test_embeddings[test_labels == 0]
embeddings_class_1 = test_embeddings[test_labels == 1]

print("Calculando distâncias intra-classe...")
dist_intra_0 = euclidean_distances(embeddings_class_0) 
dist_intra_1 = euclidean_distances(embeddings_class_1)
intra_distances = np.concatenate([
    dist_intra_0[np.triu_indices(len(dist_intra_0), k=1)],
    dist_intra_1[np.triu_indices(len(dist_intra_1), k=1)]
])

print("Calculando distâncias inter-classe...")
dist_inter = euclidean_distances(embeddings_class_0, embeddings_class_1)
inter_distances = dist_inter.flatten()

stats_intra = {'Média': np.mean(intra_distances), 'Desvio Padrão': np.std(intra_distances)}
stats_inter = {'Média': np.mean(inter_distances), 'Desvio Padrão': np.std(inter_distances)}

# --- 4. EXIBIÇÃO DA TABELA DE RESULTADOS ---
print("\n" + "="*65)
print(f"  Análise de Distâncias para o Modelo: {embeddings_file_path.split('/')[-1]}")
print("="*65)
print(f"{'Tipo de Distância':<20} | {'Média':<20} | {'Desvio Padrão':<20}")
print("-" * 65)
print(f"{'Intra-classe':<20} | {stats_intra['Média']:<20.4f} | {stats_intra['Desvio Padrão']:<20.4f}")
print(f"{'Inter-classe':<20} | {stats_inter['Média']:<20.4f} | {stats_inter['Desvio Padrão']:<20.4f}")
print("="*65)
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as T
from torch.utils.data import Dataset, DataLoader
import numpy as np
from PIL import Image
import os
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from tqdm import tqdm

# --- 1. CONFIGURAÇÃO INICIAL ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Usando o dispositivo: {device}")

IMAGE_SIZE = 128
BATCH_SIZE = 32
LOCAL_PNG_ROOT = './db/SHdataset_12k'
TEST_SPLIT_RATIO = 0.2
EMBEDDING_DIM = 64
K_NEIGHBORS = 5

# --- 2. DEFINIÇÕES NECESSÁRIAS (COPIADAS DO SCRIPT DE TREINO) ---
class EmbeddingNet(nn.Module):
    def __init__(self, embedding_dim=EMBEDDING_DIM):
        super(EmbeddingNet, self).__init__()
        self.resnet = models.resnet18(weights=None)
        num_ftrs = self.resnet.fc.in_features
        self.resnet.fc = nn.Linear(num_ftrs, embedding_dim)
    def forward(self, x):
        return self.resnet(x)

class LocalPNGDataset(Dataset):
    def __init__(self, root_dir, filenames, image_size, split='train'):
        self.root_dir = root_dir
        self.filenames = filenames
        self.split = split
        self.image_transform = T.Compose([
            T.Resize((image_size, image_size)),
            T.ToTensor(),
            T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.filenames)
    def __getitem__(self, idx):
        filename = self.filenames[idx]
        image_path = os.path.join(self.root_dir, self.split, 'images', filename)
        mask_path = os.path.join(self.root_dir, self.split, 'masks', filename)
        image = Image.open(image_path).convert('RGB')
        mask = Image.open(mask_path)
        label = 1 if np.any(np.array(mask) > 0) else 0
        return self.image_transform(image), torch.tensor(label, dtype=torch.long)

# --- 3. CARREGAMENTO DO MODELO E PREPARAÇÃO DOS DADOS ---
model = EmbeddingNet(embedding_dim=EMBEDDING_DIM).to(device)
MODEL_FILE_PATH = 'experiment_results/model_trained_on_pngs.pth'
print(f"Carregando o modelo salvo de: {MODEL_FILE_PATH}")
model.load_state_dict(torch.load(MODEL_FILE_PATH, map_location=device))
model.eval()

# Prepara a lista de arquivos de treino e de teste
all_filenames = sorted(os.listdir(os.path.join(LOCAL_PNG_ROOT, 'train', 'images')))
num_total_files = len(all_filenames)
num_test_files = int(num_total_files * TEST_SPLIT_RATIO)
num_train_files = num_total_files - num_test_files
train_filenames = all_filenames[:num_train_files]
test_filenames = all_filenames[num_train_files:]

# Cria os DataLoaders para ambos os conjuntos
train_dataset = LocalPNGDataset(LOCAL_PNG_ROOT, train_filenames, IMAGE_SIZE, split='train')
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_dataset = LocalPNGDataset(LOCAL_PNG_ROOT, test_filenames, IMAGE_SIZE, split='train')
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# --- 4. GERAÇÃO DE TODOS OS EMBEDDINGS (INFERÊNCIA) ---
def generate_embeddings(loader, desc):
    embeddings_list, labels_list = [], []
    with torch.no_grad():
        for images, labels in tqdm(loader, desc=desc):
            embeddings = model(images.to(device))
            embeddings_list.append(embeddings.cpu().numpy())
            labels_list.append(labels.numpy())
    return np.concatenate(embeddings_list), np.concatenate(labels_list)

print("\nGerando embeddings para o conjunto de TREINO...")
train_embeddings, train_labels = generate_embeddings(train_loader, "Inferência (Treino)")
print("\nGerando embeddings para o conjunto de TESTE...")
test_embeddings, test_labels = generate_embeddings(test_loader, "Inferência (Teste)")

# --- 5. AVALIAÇÃO COM K-NN (MÉTODO ORIGINAL) ---
print("\nTreinando e avaliando o k-NN com a metodologia original...")
knn = KNeighborsClassifier(n_neighbors=K_NEIGHBORS)
knn.fit(train_embeddings, train_labels) # Treina com os embeddings de treino
predictions = knn.predict(test_embeddings) # Testa com os embeddings de teste
pred_probs = knn.predict_proba(test_embeddings)[:, 1]

accuracy = accuracy_score(test_labels, predictions)
f1 = f1_score(test_labels, predictions)
auc = roc_auc_score(test_labels, pred_probs) if len(np.unique(test_labels)) > 1 else float('nan')
```

```python
print("\n" + "="*50)
print("  Métricas de Desempenho Originais (Recalculadas)")
print("="*50)
print(f"Acurácia:  {accuracy:.4f}")
print(f"F1-Score:  {f1:.4f}")
print(f"AUC Score: {auc:.4f}")
print("="*50)

# --- 6. VISUALIZAÇÃO COM T-SNE (MÉTODO ORIGINAL) ---
print("\nGerando visualização t-SNE original...")
tsne = TSNE(n_components=2, perplexity=30, init='pca', max_iter=1000, learning_rate='auto', n_jobs=-1, random_state=42)
with tqdm(total=1, desc="Executando t-SNE") as pbar:
    embeddings_2d = tsne.fit_transform(test_embeddings) # t-SNE sobre os dados de teste
    pbar.update(1)

print("Plotando o gráfico...")
plt.figure(figsize=(12, 8))

for class_id, class_name in [(0, "Negativo"), (1, "Positivo")]:
    indices = np.where(test_labels == class_id)
    plt.scatter(
        embeddings_2d[indices, 0],
        embeddings_2d[indices, 1],
        c=('blue' if class_id == 0 else 'red'),
        label=class_name,
        alpha=0.7
    )

# Rótulos dos eixos com fonte maior
plt.xlabel("t-SNE Component 1", fontsize=20)
plt.ylabel("t-SNE Component 2", fontsize=20)

# Legenda com fonte maior
plt.legend(fontsize=20)

# Ticks com fonte maior
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.grid(True)
plt.tight_layout()  # Garante que nada fique cortado
plt.show()
```
